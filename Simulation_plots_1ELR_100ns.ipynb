```python
import matplotlib.pyplot as plt

#The library which is needed to create plots - Matplotlib is imported 

#Two empty lists are created to store both 'time' and the 'Radius of gyration' values 

Time2 = []
Rg2 = []

#The .xvg file with the results is opened in read mode, and the information it contains 
#(in this case stored in two columns) is seperated and placed in two previously created lists

with open('gyrate2_1.xvg', 'r') as f:
    for line in f:
        if line.startswith(('#', '@')):
            continue
        columns = line.split()
        Time2.append(float(columns[0]))
        Rg2.append(float(columns[1]))
        
#Size of the plot, x and y axis values, labels and main title are specified and the plot is generated
        
plt.figure(figsize=(10,6))

plt.plot(Time2, Rg2)
plt.xlabel('Time (ps)')
plt.ylabel('Rg (nm)')
plt.title('Radius of gyration \n 1ELR, Unrestrained MD')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns

#Another empty list is created, this time all carboxylate clamp residues and MEEVD energy data will be stored in one list, as it will be then transformed
#into a DataFrame object, by using Pandas library

carboxylate_clamp_residues_data = []

with open('interaction_energy_single_residues.xvg', 'r') as f:
    for line in f:
        if line.startswith(('@', '#')):
            continue
        carboxylate_clamp_residues_data.append([float(data_point) for data_point in line.split()])
        
#The DataFrame is created and saved, names for each column are manually created       
        
df = pd.DataFrame(carboxylate_clamp_residues_data, columns = [ 'Time', 'Coul-SR_LYS229', 'LJ-SR_LYS229', 'Coul-SR_ASN233', 'LJ-SR_ASN233', 
                                           'Coul-SR_ASN264', 'LJ-SR_ASN264', 'Coul-SR_LYS301', 'LJ-SR_LYS301',  'Coul-SR_ARG305', 'LJ-SR_ARG305'])
```

```python
#DataFrame is then transformed by using Pandas 'melt' function, the column names from the original DataFrame are put into rows in a column
#called 'Interaction_residue', there is also column 'Time' and 'Energy_value'

df_transformed = pd.melt(df, id_vars=['Time'], var_name='Interaction_residue', value_name='Energy_value')

#The object is then transformed further by seperating the information from the 'Interaction_residue' column into two
#new columns - 'Residue_name' and 'Energy_Type'

df_transformed['Residue_name'] = df_transformed['Interaction_residue'].apply(lambda x: x.split('_')[1])
df_transformed['Energy_Type'] = df_transformed['Interaction_residue'].apply(lambda x: x.split('_')[0])

#The column 'Interaction_residue' is then removed by using 'drop' function, and data from the DataFrame is then
#seperated into two new objects - one including all Coulombic energy values and other Lennard-Jones values

df_transformed = df_transformed.drop(columns = ['Interaction_residue'])

Coul_SR_carboxylate_clamp = df_transformed[df_transformed['Energy_Type'] == 'Coul-SR']
LJ_SR_carboxylate_clamp = df_transformed[df_transformed['Energy_Type'] == 'LJ-SR']
```

```python
#To analyse and compare the distribution of both LJ-SR and Coul-SR, between all single carboxylate clamp residues
#and the MEEVD peptide, kde and scatter plots were used from seaborn library

#First for easier comparison the subplots function was used and the 'sharey = True', to ensure both plots
#share y axis and values range is the same

figure, axes = plt.subplots(1, 2, figsize=(15, 7), sharey=True)

#Four plots in total, with two for each energy type layered on top of each other. X axis values show the time frame of the
#simulation, y axis show energy values and legend describes which colour belongs to which residue
#There are also additional options used, such as choosing colours of the data points and their edge colour

#All the labels, titles and the main title are specified, along with the position of legends in both cases, and the figure is generated

sns.kdeplot(data=Coul_SR_carboxylate_clamp, x='Time', y='Energy_value', hue='Residue_name', palette='tab10', ax=axes[0], fill=True)
sns.scatterplot(data=Coul_SR_carboxylate_clamp, x='Time', y='Energy_value', hue='Residue_name', palette='tab10', ax=axes[0], alpha=0.5, edgecolor='w')
axes[0].set_title('Coul-SR')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Interaction Energy')
axes[0].set_xlim(0, 10000)

sns.kdeplot(data=LJ_SR_carboxylate_clamp, x='Time', y='Energy_value', hue='Residue_name', palette='tab10', ax=axes[1], fill=True)
sns.scatterplot(data=LJ_SR_carboxylate_clamp, x='Time', y='Energy_value', hue='Residue_name', palette='tab10', ax=axes[1], alpha=0.5, edgecolor='w')
axes[1].set_title('LJ-SR')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Interaction Energy')
axes[1].set_xlim(0, 10000)

plt.suptitle('Coulombic and Lennard-Jones Energies vs. Time \n 1ELR, carboxylate clamp residues vs. ligand')

axes[0].legend(title='Residue', bbox_to_anchor=(0.80, 0.05), loc='lower left')
axes[1].legend(title='Residue', bbox_to_anchor=(0.80, 0.05), loc='lower left')
plt.tight_layout()

plt.show()
```

```python
import pandas as pd

#Two new empty lists are created 
#One for storing energy and residues names information, second for energy values

LJ_and_COUL = []
energetics_data = []

#File is open in read mode and each of the energy labels (which include energy type and residues names) is seperated from numerical data 
#and put in seperate lists, created before.

with open('interaction_energy_heatmap_1ELR.xvg', 'r') as f:
    for line in f:
        if line.startswith('@'):
            if 'LJ-SR' in line:
                energy_names = line.strip().split('"')[1]
                LJ_and_COUL.append(f'{energy_names}')
            elif 'Coul-SR' in line:
                energy_names = line.strip().split('"')[1]
                LJ_and_COUL.append(f'{energy_names}')
        elif not line.startswith(('#', '@')):
            energetics_data.append([float(data_points) for data_points in line.split()])


#New DataFrame object is created and saved using the energy data. The column names are added using 'columns' function.

df_heatmap = pd.DataFrame(energetics_data)
df_heatmap.columns = ['time'] + LJ_and_COUL
```

```python
#DataFrame is then split into two new ones, one containing only Lennard-Jones energy information and the other Coulombic

df_LJ_heatmap = df_heatmap[[column for column in df_heatmap.columns if 'LJ-SR' in column]]
df_COUL_heatmap = df_heatmap[[column for column in df_heatmap.columns if 'Coul-SR' in column]]
```

```python
#Medians, for each column in each DataFrame, are calculated and saved

LJ_medians = df_LJ_heatmap.median()
Coul_medians = df_COUL_heatmap.median()

#Additonaly, two new DataFrame objects are created for storing the medians, with two columns - one for labels and one for values

df_LJ_medians = pd.DataFrame({
    'Residues_and_interaction': LJ_medians.index,
    'Median': LJ_medians.values
})

df_COUL_medians = pd.DataFrame({
    'Residues_and_interaction': Coul_medians.index,
    'Median': Coul_medians.values
})
```

```python
#The labels from column 'Residues_and_interaction' are then further seperated into two new columns - 'Energy_type' and 'Residues',
#and then 'Residues' are again seperated into 'Residue_1' and 'Residue_2' columns

df_LJ_medians[['Energy_type', 'Residues']] = df_LJ_medians['Residues_and_interaction'].str.split(':', expand=True)
df_LJ_medians[['Residue_1', 'Residue_2']] = df_LJ_medians['Residues'].str.split('-', expand=True)

df_COUL_medians[['Energy_type', 'Residues']] = df_COUL_medians['Residues_and_interaction'].str.split(':', expand=True)
df_COUL_medians[['Residue_1', 'Residue_2']] = df_COUL_medians['Residues'].str.split('-', expand=True)
```

```python
#Both DataFrames are then transformed by using the 'pivot_table' function. 

df_transformed_LJ = df_LJ_medians.pivot_table(index = 'Residue_1', columns = 'Residue_2', values = 'Median')

df_transformed_COUL = df_COUL_medians.pivot_table(index = 'Residue_1', columns = 'Residue_2', values = 'Median')
```

```python
#Finally, heatmaps are generated by using the heatmap function from the seaborn library. The plot sizes are
#specified, along with the colour range, main title and the labels. 

plt.figure(figsize=(10, 8))
sns.heatmap(df_transformed_LJ, annot=True, cmap = 'Blues_r')
plt.title('Medians of LJ-SR Interactions between the Hop TPR domain and Hsp90 peptide \n 1ELR')
plt.xlabel('Hsp90 ligand residues')
plt.ylabel('TPR domain residues')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_transformed_COUL, annot=True, cmap = 'Blues_r')
plt.title('Medians of Coul-SR Interactions between the Hop TPR domain and Hsp90 peptide \n 1ELR')
plt.xlabel('Hsp90 ligand residues')
plt.ylabel('TPR domain residues')
plt.show()
```
