```python
import matplotlib.pyplot as plt
import seaborn as sns

#The library which is needed to create plots - Matplotlib is imported 

#Two empty lists are created to store both 'time' and the 'Radius of gyration' values 

Time2 = []
Rg2 = []

#The .xvg file with the results is opened in read mode, and the information it contains 
#(in this case stored in two columns) is seperated and placed in two previously created lists

with open('gyrate2_1.xvg', 'r') as f:
    for line in f:
        if line.startswith(('#', '@')):
            continue
        columns = line.split()
        Time2.append(float(columns[0]))
        Rg2.append(float(columns[1]))
        
#Size of the plot, x and y axis values, labels and main title are specified and the plot is generated
        
plt.figure(figsize=(10,6))

plt.plot(Time2, Rg2)
plt.ylim(1.55, 1.68)
plt.xlabel('Time (ps)')
plt.ylabel('Rg (nm)')
plt.title('Radius of gyration \n 1ELR, Unrestrained MD')
plt.grid(True)
plt.show()
```

```python
#Three empty lists are created for storing two energy values - Lennard-Jones and Coulombic, but also time frame values of the simulation 

Time_energies2 = []
Coul_SR2 = []
LJ_SR2 = []

#The file containing energy values from the simulation is opened in read mode, and data from the three columns it contain is split in between
#the three lists previously created 

with open('interaction_energy_1ELR_clamp_ligand.xvg', 'r') as f:
    for line in f:
        if line.startswith(('@', '#')):
            continue
        columns = line.split()
        Time_energies2.append(float(columns[0]))
        Coul_SR2.append(float(columns[1]))
        LJ_SR2.append(float(columns[2]))

#The size of the plot, the x and y values (in this case for two different plots in one, as we have two different energies
# shown on the same figure), labels, title and legend are specified and the plot is generated
        
plt.figure(figsize=(10,6))
plt.plot(Time_energies2, Coul_SR2, label = 'Coulombic Energy', color = 'black')
plt.plot(Time_energies2, LJ_SR2, label = 'Lennard-Jones Energy', color = 'brown')
plt.xlabel('Time (ps)')
plt.ylabel('Energy (kJ/mol)')
plt.title('Coulombic and Lennard-Jones Energies vs. Time \n 1ELR, TPR carboxylate clamp vs. ligand')
plt.legend()
plt.grid(True)
plt.show()     
```

```python
median_Coul_SR2 = np.median(Coul_SR2)
median_LJ_SR2 = np.median(LJ_SR2)

print(f'Median of Coul-SR: {median_Coul_SR2:.2f} kJ/mol')
print(f'Median of LJ-SR: {median_LJ_SR2:.2f} kJ/mol')
```

```python
import pandas as pd

#Two new empty lists are created 
#One for storing energy and residues names information, second for energy values

LJ_and_COUL = []
energetics_data = []

#File is open in read mode and each of the energy labels (which include energy type and residues names) is seperated from numerical data 
#and put in seperate lists, created before.

with open('interaction_energy_heatmap_1ELR.xvg', 'r') as f:
    for line in f:
        if line.startswith('@'):
            if 'LJ-SR' in line:
                energy_names = line.strip().split('"')[1]
                LJ_and_COUL.append(f'{energy_names}')
            elif 'Coul-SR' in line:
                energy_names = line.strip().split('"')[1]
                LJ_and_COUL.append(f'{energy_names}')
        elif not line.startswith(('#', '@')):
            energetics_data.append([float(data_points) for data_points in line.split()])


#New DataFrame object is created and saved using the energy data. The column names are added using 'columns' function.

df_heatmap = pd.DataFrame(energetics_data)
df_heatmap.columns = ['time'] + LJ_and_COUL
```

```python
#DataFrame is then split into two new ones, one containing only Lennard-Jones energy information and the other Coulombic

df_LJ_heatmap = df_heatmap[[column for column in df_heatmap.columns if 'LJ-SR' in column]]
df_COUL_heatmap = df_heatmap[[column for column in df_heatmap.columns if 'Coul-SR' in column]]
```

```python
#Medians, for each column in each DataFrame, are calculated and saved

LJ_medians = df_LJ_heatmap.median()
Coul_medians = df_COUL_heatmap.median()

#Additonaly, two new DataFrame objects are created for storing the medians, with two columns - one for labels and one for values

df_LJ_medians = pd.DataFrame({
    'Residues_and_interaction': LJ_medians.index,
    'Median': LJ_medians.values
})

df_COUL_medians = pd.DataFrame({
    'Residues_and_interaction': Coul_medians.index,
    'Median': Coul_medians.values
})
```

```python
#The labels from column 'Residues_and_interaction' are then further seperated into two new columns - 'Energy_type' and 'Residues',
#and then 'Residues' are again seperated into 'Residue_1' and 'Residue_2' columns

df_LJ_medians[['Energy_type', 'Residues']] = df_LJ_medians['Residues_and_interaction'].str.split(':', expand=True)
df_LJ_medians[['Residue_1', 'Residue_2']] = df_LJ_medians['Residues'].str.split('-', expand=True)

df_COUL_medians[['Energy_type', 'Residues']] = df_COUL_medians['Residues_and_interaction'].str.split(':', expand=True)
df_COUL_medians[['Residue_1', 'Residue_2']] = df_COUL_medians['Residues'].str.split('-', expand=True)
```

```python
#Both DataFrames are then transformed by using the 'pivot_table' function. 

df_transformed_LJ = df_LJ_medians.pivot_table(index = 'Residue_1', columns = 'Residue_2', values = 'Median')

df_transformed_COUL = df_COUL_medians.pivot_table(index = 'Residue_1', columns = 'Residue_2', values = 'Median')
```

```python
#Finally, heatmaps are generated by using the heatmap function from the seaborn library. The plot sizes are
#specified, along with the colour range, main title and the labels. 

plt.figure(figsize=(10, 8))
sns.heatmap(df_transformed_LJ, annot=True, vmin = -10, vmax = 0, cmap = 'Blues_r')
plt.title('Medians of LJ-SR Interactions between the Hop TPR domain and Hsp90 peptide \n 1ELR')
plt.xlabel('Hsp90 ligand residues')
plt.ylabel('TPR domain residues')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_transformed_COUL, annot=True, vmin = -130, vmax = 0, cmap = 'Blues_r')
plt.title('Medians of Coul-SR Interactions between the Hop TPR domain and Hsp90 peptide \n 1ELR')
plt.xlabel('Hsp90 ligand residues')
plt.ylabel('TPR domain residues')
plt.show()
```
